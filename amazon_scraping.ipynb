```python
                    urls=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")

                    for i in urls[2:24]:
                         link1.append(i.get_attribute('href'))
                    for url in link1:
                        driver.get(url)

    
                        try:
                            Task1 =driver.find_element_by_xpath("//h1[@class='a-size-large a-spacing-none']/span[1]")
                            product_title1.append(Task1.text)
                        except:
                            product_title1.append(' ')

                        try:
                            Task2 =driver.find_element_by_xpath("//td[@class='a-span12']/span[1]")
                            product_price1.append(Task2.text)
                        except:
                            product_price1.append(' ')
                        

                        try:
                            Task3 =driver.find_element_by_xpath("//div[@class='celwidget']/div[@class='a-section a-spacing-none']/a[1]")
                            brand1.append(Task3.text)
                        except:
                            brand1.append(' ')
                        
                        
                        try:
                            Task4 =driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']")
                            Product_Description1.append(Task4.text)
                        except:
                            Product_Description1.append(' ')

                        
```
